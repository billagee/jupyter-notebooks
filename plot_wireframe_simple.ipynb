```python
%matplotlib inline

# See http://matplotlib.org/examples/mplot3d/wire3d_demo.html

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create 2d array for X
x_row = np.arange(4)       # array([0, 1, 2, 3])
X = np.tile(x_row, (4, 1)) # repeat x_row 4 times
Y = np.rot90(X, 3)  # Y is X rotated 270 degrees counter-clockwise
Z = np.ones((4, 4)) # 4x4 2d array of 1s

print(X) ; print(Y) ; print(Z)
ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1)

plt.title('Foo')
plt.xlabel("Here's X")
plt.ylabel("Here's Y")
plt.show()
```
